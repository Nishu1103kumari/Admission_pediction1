```python
import pandas as pd
df = pd.read_csv('Admission_Predict.csv')
df.head(2)
```

```python
y = df['Chance of Admit ']
x = df[['GRE Score', 'University Rating', 'SOP',
       'LOR ', 'CGPA', 'Research']]
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.3,random_state = 9598)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error,mean_squared_error
mean_absolute_error(y_test,y_pred)
```

```python
per_e = mean_absolute_percentage_error(y_test,y_pred)
per_e
```

```python
accuracy1 = (1-per_e)*100
accuracy1
```

```python
mean_squared_error(y_test,y_pred)
```

```python
from sklearn.neighbors import KNeighborsRegressor
model2 = KNeighborsRegressor() 
model2.fit(x_train,y_train)
```

```python
y_pred2 = model2.predict(x_test)
y_pred2[:5]
```

```python
y_test[:5]
```

```python
error2 = mean_absolute_percentage_error(y_test,y_pred2)
error2
```

```python
acc2 = (1-error2)*100
acc2
```

```python
from sklearn.tree import DecisionTreeRegressor
model3 = DecisionTreeRegressor()
model3.fit(x_train,y_train)
```

```python
y_pred3 = model3.predict(x_test)
y_pred3[:5]
```

```python
y_test[:5]
```

```python
error3 = mean_absolute_percentage_error(y_test,y_pred3)
error3
```

```python
acc3 = (1-error3)*100
acc3
```
